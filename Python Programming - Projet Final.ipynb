```python
#Import des variables depuis fichier excel dans un dictionnaire (partie 1)
import pandas as pd

#METTRE LE PATH DES BILANS-RETRAITES ICI (FICHIER: bilans_retraites.xlsx)
excel_file_path = r"/Users/benoitchedru/Documents/School/COURSES/PythonProgramming/ProjetFinal/bilans_retraites.xlsx"

df_bilan = pd.read_excel(excel_file_path, header=None)

key_enterprises = df_bilan.iloc[0, 0]


financial_info_names = [
    'Cash And Cash Equivalents',
    'Common Stock Equity',
    'Net Debt',
    'Ordinary Shares Number',
    'Share Issued',
    'Total debt'
]

#dictionaire principal
data_dict_bilan = {key_enterprises: {}}

for index, row in df_bilan.iterrows():

    if index == 0:
        continue
    enterprise_name = row.iloc[0]
    enterprise_dict = {ratio_name: value for ratio_name, value in zip(financial_info_names, row[1:])}
    data_dict_bilan[key_enterprises][enterprise_name] = enterprise_dict

print(data_dict_bilan)
```

```python
#Import des variables depuis fichier excel dans un [autre] dictionnaire (partie 2) - c'est le m√™me proc√©d√© que la boite pr√©c√©dente

import pandas as pd


#METTRE LE PATH DES CDR-RETRAITES ICI (FICHIER: cdr_retraites.xlsx)
excel_file_path = r"/Users/benoitchedru/Documents/School/COURSES/PythonProgramming/ProjetFinal/cdr_retraites.xlsx"


df_cdr = pd.read_excel(excel_file_path, header=None)


financial_info_names = [
    'Basic EPS',
    'Diluted EPS',
    'Diluted NI Availto Com Stockholders',
    'Net Income',
    'Operating Income',
    'Total Revenue'
]


data_dict_cdr = {key_enterprises: {}}

for index, row in df_cdr.iterrows():
    
    if index == 0:
        continue
    enterprise_name = row.iloc[0]   
    enterprise_dict = {ratio_name: value for ratio_name, value in zip(financial_info_names, row[1:])}
    data_dict_cdr[key_enterprises][enterprise_name] = enterprise_dict


print(data_dict_cdr)
```

```python
# Cr√©ation du dictionnaire principal avec toutes les variables "allratios"
# Nous allons faire √ßa en cr√©ant un dictionnaire principal contenant les informations du compte de profits et pertes 
# puis en utilisant une boucle for pour ajouter toutes les informations du bilan une par une

allratios=data_dict_cdr

for company in entreprises:
    #certaines variables n'ont pas pu √™tre trouv√©es; il faut √™tre sur de pouvoir les afficher comme "not available (non-disponible)"
    if len(allratios['Companies'][company])==0:
        allratios['Companies'][company]['empty']='empty'
    else:
        for balancesheet_element in data_dict_bilan['Companies'][company]:
            try:
                allratios['Companies'][company][balancesheet_element]=data_dict_bilan['Companies'][company][balancesheet_element]
            except: 
                allratios['Companies'][company][balancesheet_element]='not available'
            else:
                allratios['Companies'][company][balancesheet_element]=data_dict_bilan['Companies'][company][balancesheet_element]

print(allratios)
```

```python
#pour pouvoir filtrer toutes les variables, nous combinons le tout dans le dictionnaire principal

variables=allratios['Companies']['LVMH']
#naturellement, on veut que les keys et pas que les valeurs de LVMH, par exemple

keys_list = list(variables.keys())

print(keys_list)
```
